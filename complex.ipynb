```python
from math import sqrt, pi, atan, cos,asin, acos,sin

class C_numb:
    def __init__(self,real=1,im=0):
        self.mode='norm'
        self.real = real
        self.im = im
        self.mod = sqrt(real**2 + im**2)

        if real>0 and im>=0:
            self.arg = atan(im/real)

        elif real >0 and im<0:
            self.arg = atan(im/real) + 2*pi

        elif real<0 and im>=0:
            self.arg = atan(im/real) + pi

        elif real<0 and im<0:
            self.arg = atan(im/real) + pi

        elif real == 0:
            if im<0:
                self.arg = 2*pi - pi*0.5

            elif im>0:
                self.arg = pi*0.5

    def __add__(self,a):
        if isinstance(a,C_numb):
            return C_numb(self.real + a.real,self.im + a.im)
        else:
            return C_numb(self.real + a,self.im)

    def __radd__(self,a):
        if isinstance(a,C_numb):
            return C_numb(self.real + a.real,self.im + a.im)
        else:
            return C_numb(self.real + a,self.im)

    def __mul__(self,a):
        if isinstance(a,C_numb):
            return C_numb(self.real*a.real - self.im*a.im,self.real*a.im + self.im*a.real)
        else:
            return C_numb(self.real*a,self.im*a)

    def __rmul__(self,a):
        if isinstance(a,C_numb):
            return C_numb(self.real*a.real - self.im*a.im,self.real*a.im + self.im*a.real)
        else:
            return C_numb(self.real*a,self.im*a)

    def con(self):
        return C_numb(self.real,-1*self.im)

    def __truediv__(self,a):
        if isinstance(a,C_numb):
            num = self*a.con()
            denom = (a*a.con()).real
            result = num*(denom**(-1))
            return result
        else:
            return self*(a**(-1))
    def __sub__(self, a):
        if isinstance(a,C_numb):
             return C_numb(self.real - a.real,self.im - a.im)
        else:
            return C_numb(self.real - a,self.im)
    def to_trig(self):
        self.mode='trig'
    def __pow__(self,n):
        t=C_numb(0,0)
        t.mode='trig'
        t.mod=self.mod**n
        t.arg=(n*self.arg)
        return(t)
    def sq(self,n):
        n=int(n)
        p=[]
        if self.mode == 'norm':
            self.to_trig()
        for i in range(0,n):
            p.append(C_numb(0,0))
            p[i].mode='trig'
            p[i].mod= self.mod**(1/n)
            p[i].arg = (self.arg+2*pi*i)/n
        return(p)
    def mul(self, a):
        q=C_numb(0,0)
        q.mode='trig'
        q.mod=self.mod*a.mod
        q.arg=self.arg+a.arg
        return q
    def __str__(self):
        if self.mode=='norm':
            return '{} + i*{}'.format(self.real, self.im)
        else:
            return '|{}|*({}+i*{})'.format(self.mod, cos(self.arg), sin(self.arg))
a=C_numb(3,2)
b=C_numb(3,2)
a=a.mul(b)
if isinstance(a,list):
    for i in range(len(a)):
        print(a[i])
else:
    print(a)
```
